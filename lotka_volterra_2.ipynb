**Modelo Lotka-Volterra: Interacción Presa-Depredador**

**Introducción**
El modelo Lotka-Volterra es un sistema de ecuaciones diferenciales que describe la dinámica de poblaciones de dos especies: una población de presas (ej. conejos) y una de depredadores (ej. lobos). Este modelo es fundamental para entender cómo dos especies interactúan en un ecosistema, particularmente cuando una depende de la otra para su alimentación.

**1. Supuestos del Modelo**

Antes de abordar las ecuaciones, necesitamos entender algunos supuestos básicos:

•	**Crecimiento exponencial de la presa:** En ausencia de depredadores, la población de presas crece exponencialmente, es decir, sigue aumentando sin límite, ya que tienen comida abundante y espacio suficiente.

•	**Dependencia de la presa para los depredadores:** Los depredadores dependen de las presas para sobrevivir. Si no hay presas, los depredadores mueren.

•	**Tasa de captura proporcional:** La tasa a la que los depredadores capturan presas es proporcional a la cantidad de interacción entre ellos.

La función lotka_volterra representa el modelo de depredador-presa y utiliza los siguientes parámetros:

**y:** Es un vector (o lista) que contiene el estado actual del sistema. En este caso, se asume que y[0] corresponde a la población de conejos (presas) y y[1] a la población de lobos (depredadores).

**t:** Representa el tiempo. Aunque en este modelo los parámetros no varían explícitamente con el tiempo, este parámetro es necesario para que funciones como odeint puedan integrar la ecuación diferencial a lo largo de un intervalo temporal.

**alpha (α):** Tasa de crecimiento natural de la población de conejos en ausencia de depredadores. Este parámetro indica cómo se multiplican los conejos cuando no hay lobos.

**beta (β):** Coeficiente de depredación. Representa la probabilidad o la tasa con la que los lobos cazan a los conejos. Se multiplica por el producto de las poblaciones de conejos y lobos para modelar el número de conejos que son cazados.

**gamma (γ):** Tasa de crecimiento de los lobos como resultado del consumo de conejos. Este parámetro traduce cuántos lobos "nacen" o cuánta energía se obtiene para la supervivencia y reproducción de los lobos a partir de cada conejo cazado.

**delta (δ):** Tasa de mortalidad de los lobos. Indica la proporción de lobos que mueren por causas naturales o por la falta de alimento en ausencia de suficientes conejos.


```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
import ipywidgets as widgets
from ipywidgets import interact

# Definición del modelo Lotka-Volterra
def lotka_volterra(y, t, alpha, beta, gamma, delta):
    conejos, lobos = y
    dConejos_dt = alpha * conejos - beta * conejos * lobos
    dLobos_dt = gamma * conejos * lobos - delta * lobos
    return [dConejos_dt, dLobos_dt]

# Función que simula y grafica el modelo
def simular_lotka_volterra(alpha, beta, gamma, delta):
    t = np.linspace(0, 200, 1000)
    y0 = [40, 9]  # Condiciones iniciales: [Conejos, Lobos]
    sol = odeint(lotka_volterra, y0, t, args=(alpha, beta, gamma, delta))
    
    plt.figure(figsize=(10,6))
    plt.plot(t, sol[:, 0], 'g', label='Conejos')
    plt.plot(t, sol[:, 1], 'r', label='Lobos')
    plt.xlabel('Tiempo')
    plt.ylabel('Población')
    plt.title('Simulación interactiva del modelo Lotka-Volterra')
    plt.legend()
    plt.show()

# Creación de sliders interactivos para cada parámetro
interact(simular_lotka_volterra,
         alpha=widgets.FloatSlider(min=0.01, max=0.2, step=0.01, value=0.1, description='α'),
         beta=widgets.FloatSlider(min=0.001, max=0.1, step=0.001, value=0.02, description='β'),
         gamma=widgets.FloatSlider(min=0.001, max=0.05, step=0.001, value=0.01, description='γ'),
         delta=widgets.FloatSlider(min=0.01, max=1.0, step=0.01, value=0.1, description='δ'))
```

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.integrate import odeint
import ipywidgets as widgets
from ipywidgets import interactive, Tab, VBox, fixed
from IPython.display import display

# Definición del modelo Lotka-Volterra
def lotka_volterra(y, t, alpha, beta, gamma, delta):
    conejos, lobos = y
    dConejos_dt = alpha * conejos - beta * conejos * lobos
    dLobos_dt = gamma * conejos * lobos - delta * lobos
    return [dConejos_dt, dLobos_dt]

# Función para simular y graficar el modelo según parámetros y título
def simular_scenario(alpha, beta, gamma, delta, title):
    t = np.linspace(0, 200, 1000)
    y0 = [40, 9]  # Condiciones iniciales: [Conejos, Lobos]
    sol = odeint(lotka_volterra, y0, t, args=(alpha, beta, gamma, delta))
    
    plt.figure(figsize=(8, 4))
    plt.plot(t, sol[:, 0], 'g', label='Conejos')
    plt.plot(t, sol[:, 1], 'r', label='Lobos')
    plt.xlabel('Tiempo')
    plt.ylabel('Población')
    plt.title(title)
    plt.legend()
    plt.show()

# Pestaña 1: Escenario 1 - Equilibrio
interactivo_eq = interactive(simular_scenario,
    alpha=widgets.FloatSlider(min=0.01, max=0.2, step=0.01, value=0.1, description='α'),
    beta=widgets.FloatSlider(min=0.005, max=0.1, step=0.005, value=0.02, description='β'),
    gamma=widgets.FloatSlider(min=0.001, max=0.05, step=0.001, value=0.01, description='γ'),
    delta=widgets.FloatSlider(min=0.01, max=0.5, step=0.01, value=0.1, description='δ'),
    title=fixed('Escenario 1: Equilibrio')
)

# Pestaña 2: Escenario 2 - Extinción de lobos
interactivo_lobos_ext = interactive(simular_scenario,
    alpha=widgets.FloatSlider(min=0.01, max=0.2, step=0.01, value=0.1, description='α'),
    beta=widgets.FloatSlider(min=0.005, max=0.1, step=0.005, value=0.02, description='β'),
    gamma=widgets.FloatSlider(min=0.0001, max=0.05, step=0.0005, value=0.001, description='γ'),
    delta=widgets.FloatSlider(min=0.1, max=1.0, step=0.1, value=0.8, description='δ'),
    title=fixed('Escenario 2: Extinción de lobos')
)

# Pestaña 3: Escenario 3 - Extinción de conejos
interactivo_conejos_ext = interactive(simular_scenario,
    alpha=widgets.FloatSlider(min=0.01, max=0.1, step=0.005, value=0.05, description='α'),
    beta=widgets.FloatSlider(min=0.01, max=0.2, step=0.01, value=0.1, description='β'),
    gamma=widgets.FloatSlider(min=0.001, max=0.05, step=0.001, value=0.02, description='γ'),
    delta=widgets.FloatSlider(min=0.1, max=0.5, step=0.01, value=0.2, description='δ'),
    title=fixed('Escenario 3: Extinción de conejos')
)

# Crear pestañas para organizar los escenarios
tab = Tab(children=[interactivo_eq, interactivo_lobos_ext, interactivo_conejos_ext])
tab.set_title(0, 'Equilibrio')
tab.set_title(1, 'Extinción de lobos')
tab.set_title(2, 'Extinción de conejos')

display(tab)
```

**Escenario de equilibrio:** Tanto conejos como lobos se estabilizan, oscilando alrededor de valores equilibrados sin que ninguna de las dos especies se extinga.

**Extinción de lobos:** La población de lobos disminuye rápidamente debido a una tasa de mortalidad alta, mientras que la población de conejos crece sin control en ausencia de depredadores.

**Extinción de conejos:** Los conejos desaparecen rápidamente debido a una tasa de depredación alta, lo que lleva a la eventual caída de la población de lobos por falta de presas. ​
